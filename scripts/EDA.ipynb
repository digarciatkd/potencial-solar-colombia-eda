```python
# %% [markdown]
# # EDA — Potencial Solar Colombia vs. Proyectos UPME (2024–2026)
#
# **Pregunta de investigación:**
# ¿Existe correlación entre las zonas con mayor irradiancia solar histórica en Colombia
# (2024–2026) y la distribución de proyectos fotovoltaicos registrados en la UPME?
# ¿Qué regiones muestran la mayor brecha entre recurso disponible y planificación actual?
#
# **Fuentes de datos:**
# - NASA POWER API: `ALLSKY_SFC_SW_DWN`, `T2M`, `RH2M`, `WS2M`, `PRECTOTCORR`
#   para 88 municipios representativos de las cinco regiones naturales de Colombia.
# - Registro oficial de proyectos solares UPME (72 proyectos fotovoltaicos, 7 155 MW).
#
# **Período:** 2024-01-01 → 2026-05-09 (860 días, 74 984 mediciones diarias)

# %% [markdown]
# ## 0 · Configuración e importaciones

# %%
import sqlite3
import warnings
from pathlib import Path

import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import matplotlib.patches as mpatches
import matplotlib.gridspec as gridspec
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
from scipy.stats import pearsonr, spearmanr, kendalltau

warnings.filterwarnings("ignore")

# ── Estética global ──────────────────────────────────────────────────────────
PALETTE_REGION = {
    "Caribe":    "#F4A020",
    "Andina":    "#3A86FF",
    "Orinoquia": "#06D6A0",
    "Amazonia":  "#8338EC",
    "Pacifica":  "#FF006E",
    "Insular":   "#FB5607",
}
SOLAR_ORANGE = "#F4A020"
UPME_BLUE    = "#3A86FF"
GAP_RED      = "#EF233C"

sns.set_theme(style="whitegrid", font_scale=1.1)
plt.rcParams.update({
    "figure.facecolor": "#FAFAFA",
    "axes.facecolor":   "#FAFAFA",
    "axes.edgecolor":   "#CCCCCC",
    "grid.color":       "#E8E8E8",
    "font.family":      "DejaVu Sans",
})

# ── Ruta a la base de datos ──────────────────────────────────────────────────
DB_PATH = Path("potencial_solar.db")   # ← ajustar si el archivo está en otra carpeta
assert DB_PATH.exists(), f"No se encontró la base de datos en: {DB_PATH.resolve()}"
print(f"✅  Base de datos encontrada: {DB_PATH.resolve()}")

# %% [markdown]
# ## 1 · Carga de datos

# %%
conn = sqlite3.connect(DB_PATH)

# ── 1.1 Dimensiones de referencia ────────────────────────────────────────────
df_departamentos = pd.read_sql("SELECT * FROM departamentos", conn)
df_municipios    = pd.read_sql("""
    SELECT m.id_municipio, m.nombre municipio, m.latitud, m.longitud,
           d.nombre departamento, d.region_natural, d.codigo_dane codigo_dpto
    FROM municipios m
    JOIN departamentos d USING(id_departamento)
""", conn)

df_recursos        = pd.read_sql("SELECT * FROM recursos", conn)
df_tecnologias     = pd.read_sql("SELECT * FROM tecnologias", conn)
df_estados         = pd.read_sql("SELECT * FROM estados_proyecto", conn)

# ── 1.2 Mediciones NASA ───────────────────────────────────────────────────────
df_nasa = pd.read_sql("""
    SELECT n.id_medicion, n.id_municipio, n.fecha,
           n.irradiancia_global, n.temperatura_2m, n.humedad_relativa,
           n.velocidad_viento, n.precipitacion,
           m.municipio, m.departamento, m.region_natural,
           m.latitud, m.longitud
    FROM mediciones_nasa n
    JOIN (
        SELECT m2.id_municipio, m2.nombre municipio, m2.latitud, m2.longitud,
               d.nombre departamento, d.region_natural
        FROM municipios m2 JOIN departamentos d USING(id_departamento)
    ) m USING(id_municipio)
""", conn)
df_nasa["fecha"] = pd.to_datetime(df_nasa["fecha"])
df_nasa["anio"]  = df_nasa["fecha"].dt.year
df_nasa["mes"]   = df_nasa["fecha"].dt.month
df_nasa["anio_mes"] = df_nasa["fecha"].dt.to_period("M")

# ── 1.3 Proyectos UPME (todos los recursos) ───────────────────────────────────
df_upme_all = pd.read_sql("""
    SELECT p.*,
           ep.nombre estado,
           r.nombre  recurso,
           t.nombre  tecnologia,
           m.nombre  municipio,
           m.latitud, m.longitud,
           d.nombre  departamento,
           d.region_natural
    FROM proyectos_upme p
    JOIN estados_proyecto ep USING(id_estado)
    JOIN recursos          r  USING(id_recurso)
    JOIN tecnologias        t  USING(id_tecnologia)
    JOIN municipios         m  USING(id_municipio)
    JOIN departamentos      d  ON m.id_departamento = d.id_departamento
""", conn)
for col in ["fecha_inscripcion","fecha_validez",
            "fecha_inicio_construccion","fecha_entrada_operacion"]:
    df_upme_all[col] = pd.to_datetime(df_upme_all[col])

# ── 1.4 Sólo proyectos SOLARES / FOTOVOLTAICOS ───────────────────────────────
df_solar = df_upme_all[df_upme_all["recurso"] == "SOLAR"].copy()

conn.close()

print(f"\n📊  mediciones_nasa : {df_nasa.shape[0]:,} filas × {df_nasa.shape[1]} columnas")
print(f"📊  proyectos UPME  : {df_upme_all.shape[0]} filas  ({df_solar.shape[0]} solares)")
print(f"📅  Rango temporal  : {df_nasa.fecha.min().date()} → {df_nasa.fecha.max().date()}")

# %% [markdown]
# ## 2 · Calidad de datos

# %%
print("=" * 60)
print("CALIDAD DE DATOS — mediciones_nasa")
print("=" * 60)

total = len(df_nasa)
nulls = df_nasa[["irradiancia_global","temperatura_2m","humedad_relativa",
                  "velocidad_viento","precipitacion"]].isnull().sum()
print(f"\nTotal de registros : {total:,}")
print("\nValores nulos por variable:")
for var, n in nulls.items():
    pct = 100 * n / total
    flag = "⚠️" if pct > 1 else "✅"
    print(f"  {flag}  {var:<25s}: {n:4d}  ({pct:.2f}%)")

# Cobertura por municipio
coverage = (df_nasa.groupby("id_municipio")["fecha"]
                   .count()
                   .rename("n_obs"))
print(f"\nRegistros por municipio — min:{coverage.min()}  max:{coverage.max()}  "
      f"mediana:{coverage.median():.0f}")

# Duplicados
dup = df_nasa.duplicated(subset=["id_municipio","fecha"]).sum()
print(f"\nPares (municipio, fecha) duplicados: {dup}")

# Outliers irradiancia (valores físicamente imposibles)
irr_hi = (df_nasa["irradiancia_global"] > 10).sum()
irr_lo = (df_nasa["irradiancia_global"] < 0).sum()
print(f"\nValores irradiancia > 10 kWh/m²/día : {irr_hi}")
print(f"Valores irradiancia <  0            : {irr_lo}")

# Imputar nulos de irradiancia con la mediana del municipio
df_nasa["irradiancia_global"] = (
    df_nasa.groupby("id_municipio")["irradiancia_global"]
           .transform(lambda s: s.fillna(s.median()))
)
print(f"\n✅  Nulos de irradiancia imputados con mediana por municipio.")

# %% [markdown]
# ## 3 · Estadísticas descriptivas — Variables NASA

# %%
vars_nasa = ["irradiancia_global","temperatura_2m","humedad_relativa",
             "velocidad_viento","precipitacion"]
labels_nasa = {
    "irradiancia_global": "Irradiancia global (kWh/m²/día)",
    "temperatura_2m"    : "Temperatura 2 m (°C)",
    "humedad_relativa"  : "Humedad relativa (%)",
    "velocidad_viento"  : "Velocidad viento 2 m (m/s)",
    "precipitacion"     : "Precipitación corregida (mm/día)",
}

desc = df_nasa[vars_nasa].describe().T
desc.index = [labels_nasa[v] for v in desc.index]
print("\nResumen estadístico — todas las mediciones:")
print(desc.round(3).to_string())

# ── Por región natural ────────────────────────────────────────────────────────
print("\n\nMedia de variables NASA por región natural:")
reg_stats = (df_nasa.groupby("region_natural")[vars_nasa]
                     .mean()
                     .round(3)
                     .sort_values("irradiancia_global", ascending=False))
print(reg_stats.to_string())

# %% [markdown]
# ## 4 · Distribución de irradiancia — histogramas y densidades

# %%
fig, axes = plt.subplots(2, 3, figsize=(16, 9))
fig.suptitle("Distribución de irradiancia global por región natural\n(kWh/m²/día — diaria)",
             fontsize=14, fontweight="bold", y=1.01)

regiones = sorted(df_nasa["region_natural"].unique())

for ax, region in zip(axes.flat, regiones):
    data = df_nasa.loc[df_nasa["region_natural"] == region, "irradiancia_global"].dropna()
    color = PALETTE_REGION.get(region, "#888888")
    ax.hist(data, bins=60, color=color, alpha=0.7, edgecolor="white", linewidth=0.3)
    ax.axvline(data.mean(),   color="#333333", lw=1.5, ls="--", label=f"Media: {data.mean():.2f}")
    ax.axvline(data.median(), color="#999999", lw=1.2, ls=":",  label=f"Mediana: {data.median():.2f}")
    ax.set_title(region, fontweight="bold", color=color)
    ax.set_xlabel("kWh/m²/día")
    ax.set_ylabel("Frecuencia")
    ax.legend(fontsize=8)
    skew = data.skew()
    ax.text(0.98, 0.95, f"n={len(data):,}\nasim={skew:.2f}",
            transform=ax.transAxes, ha="right", va="top", fontsize=8,
            bbox=dict(boxstyle="round,pad=0.3", fc="white", ec=color, alpha=0.7))

axes.flat[-1].set_visible(False)   # Insular tiene muy pocos datos → la muestra igual
# Volver a mostrar Insular en la última celda
ax_ins = axes[1, 2]
ax_ins.set_visible(True)
data_ins = df_nasa.loc[df_nasa["region_natural"] == "Insular", "irradiancia_global"].dropna()
ax_ins.hist(data_ins, bins=40, color=PALETTE_REGION["Insular"], alpha=0.7,
            edgecolor="white", linewidth=0.3)
ax_ins.axvline(data_ins.mean(), color="#333333", lw=1.5, ls="--",
               label=f"Media: {data_ins.mean():.2f}")
ax_ins.set_title("Insular", fontweight="bold", color=PALETTE_REGION["Insular"])
ax_ins.set_xlabel("kWh/m²/día")
ax_ins.set_ylabel("Frecuencia")
ax_ins.legend(fontsize=8)

plt.tight_layout()
plt.savefig("fig01_distribuciones_region.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅  Figura 1 guardada.")

# %% [markdown]
# ## 5 · Ranking municipal de irradiancia media

# %%
irr_muni = (df_nasa.groupby(["id_municipio","municipio","departamento","region_natural",
                               "latitud","longitud"])
                    ["irradiancia_global"]
                    .agg(["mean","std","count"])
                    .reset_index()
                    .rename(columns={"mean":"irr_media","std":"irr_std","count":"n_obs"}))
irr_muni = irr_muni.sort_values("irr_media", ascending=False).reset_index(drop=True)
irr_muni["rank_irr"] = irr_muni.index + 1

print(f"Municipios con medición de irradiancia: {len(irr_muni)}")
print("\nTop 15 municipios por irradiancia media:")
print(irr_muni[["rank_irr","municipio","departamento","region_natural","irr_media","irr_std"]]
      .head(15).to_string(index=False))
print("\nBottom 10 municipios por irradiancia media:")
print(irr_muni[["rank_irr","municipio","departamento","region_natural","irr_media"]]
      .tail(10).to_string(index=False))

# ── Gráfico horizontal ranking top-30 ────────────────────────────────────────
top30 = irr_muni.head(30)
fig, ax = plt.subplots(figsize=(11, 10))
colors = [PALETTE_REGION.get(r, "#888") for r in top30["region_natural"]]
bars = ax.barh(range(len(top30)), top30["irr_media"], color=colors, edgecolor="white")
ax.errorbar(top30["irr_media"], range(len(top30)),
            xerr=top30["irr_std"], fmt="none", color="#555", lw=0.7, capsize=2)
ax.set_yticks(range(len(top30)))
ax.set_yticklabels([f"{row.municipio}\n({row.departamento})"
                    for _, row in top30.iterrows()], fontsize=8)
ax.invert_yaxis()
ax.set_xlabel("Irradiancia media diaria (kWh/m²/día)")
ax.set_title("Top 30 municipios — Irradiancia solar media\n(2024–2026)", fontweight="bold")
legend_patches = [mpatches.Patch(color=c, label=r)
                  for r, c in PALETTE_REGION.items() if r in top30["region_natural"].values]
ax.legend(handles=legend_patches, loc="lower right", fontsize=9)
ax.axvline(irr_muni["irr_media"].mean(), color="#555", ls="--", lw=1,
           label=f"Media nacional: {irr_muni['irr_media'].mean():.2f}")
plt.tight_layout()
plt.savefig("fig02_ranking_municipal_irradiancia.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅  Figura 2 guardada.")

# %% [markdown]
# ## 6 · Análisis de la irradiancia por región — boxplot comparativo

# %%
order_regions = (irr_muni.groupby("region_natural")["irr_media"]
                          .median()
                          .sort_values(ascending=False)
                          .index.tolist())

fig, ax = plt.subplots(figsize=(13, 6))
bp = sns.boxplot(data=df_nasa.dropna(subset=["irradiancia_global"]),
                 x="region_natural", y="irradiancia_global",
                 order=order_regions,
                 palette=PALETTE_REGION,
                 width=0.55, linewidth=0.9, fliersize=1.5, ax=ax)
ax.set_xlabel("Región natural")
ax.set_ylabel("Irradiancia global diaria (kWh/m²/día)")
ax.set_title("Distribución de irradiancia solar por región natural (2024–2026)",
             fontweight="bold")

# Anotar medias
medias = df_nasa.groupby("region_natural")["irradiancia_global"].mean()
for i, region in enumerate(order_regions):
    m = medias[region]
    ax.text(i, m + 0.06, f"{m:.2f}", ha="center", va="bottom", fontsize=9,
            fontweight="bold", color=PALETTE_REGION.get(region, "#333"))

plt.tight_layout()
plt.savefig("fig03_boxplot_region.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅  Figura 3 guardada.")

# %% [markdown]
# ## 7 · Estacionalidad — irradiancia y precipitación a lo largo del tiempo

# %%
monthly = (df_nasa.groupby("anio_mes")
                  .agg(irr=("irradiancia_global","mean"),
                       prec=("precipitacion","mean"))
                  .reset_index())
monthly["fecha_plot"] = monthly["anio_mes"].dt.to_timestamp()

fig, ax1 = plt.subplots(figsize=(14, 5))
ax2 = ax1.twinx()

ax1.plot(monthly["fecha_plot"], monthly["irr"], color=SOLAR_ORANGE,
         lw=2.5, marker="o", ms=4, label="Irradiancia media (kWh/m²/día)")
ax1.fill_between(monthly["fecha_plot"], monthly["irr"], alpha=0.15, color=SOLAR_ORANGE)
ax2.bar(monthly["fecha_plot"], monthly["prec"], width=20,
        color=UPME_BLUE, alpha=0.4, label="Precipitación media (mm/día)")

ax1.set_ylabel("Irradiancia (kWh/m²/día)", color=SOLAR_ORANGE)
ax2.set_ylabel("Precipitación (mm/día)", color=UPME_BLUE)
ax1.tick_params(axis="y", colors=SOLAR_ORANGE)
ax2.tick_params(axis="y", colors=UPME_BLUE)
ax1.set_title("Estacionalidad nacional: Irradiancia vs. Precipitación (2024–2026)",
              fontweight="bold")

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper right")
plt.tight_layout()
plt.savefig("fig04_estacionalidad.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅  Figura 4 guardada.")

# ── Por región ────────────────────────────────────────────────────────────────
monthly_reg = (df_nasa.groupby(["anio_mes","region_natural"])
                       ["irradiancia_global"].mean()
                       .reset_index())
monthly_reg["fecha_plot"] = monthly_reg["anio_mes"].dt.to_timestamp()

fig, ax = plt.subplots(figsize=(14, 5))
for region in order_regions:
    sub = monthly_reg[monthly_reg["region_natural"] == region]
    ax.plot(sub["fecha_plot"], sub["irradiancia_global"],
            color=PALETTE_REGION.get(region, "#888"), lw=2, label=region)
ax.set_ylabel("Irradiancia media mensual (kWh/m²/día)")
ax.set_title("Evolución mensual de irradiancia por región natural (2024–2026)",
             fontweight="bold")
ax.legend(loc="lower right")
plt.tight_layout()
plt.savefig("fig05_estacionalidad_region.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅  Figura 5 guardada.")

# %% [markdown]
# ## 8 · Correlaciones entre variables climatológicas

# %%
corr_vars = vars_nasa
corr_labels = [labels_nasa[v] for v in corr_vars]

# Pearson sobre muestra aleatoria (velocidad) ─────────────────────────────────
sample = df_nasa[corr_vars].dropna().sample(min(20_000, len(df_nasa)), random_state=42)
corr_matrix = sample.corr(method="pearson")
corr_matrix.columns = corr_labels
corr_matrix.index   = corr_labels

fig, ax = plt.subplots(figsize=(9, 7))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool), k=1)
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="RdYlGn",
            center=0, vmin=-1, vmax=1,
            linewidths=0.5, ax=ax, square=True)
ax.set_title("Matriz de correlación de Pearson — Variables NASA\n(muestra aleatoria)",
             fontweight="bold")
plt.tight_layout()
plt.savefig("fig06_correlacion_nasa.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅  Figura 6 guardada.")

# Pairplot irradiancia vs resto ───────────────────────────────────────────────
print("\nCorrelaciones de Pearson con irradiancia_global:")
for var in [v for v in corr_vars if v != "irradiancia_global"]:
    sub = df_nasa[["irradiancia_global", var]].dropna()
    r, p = pearsonr(sub["irradiancia_global"], sub[var])
    sig = "***" if p < 0.001 else ("**" if p < 0.01 else ("*" if p < 0.05 else "ns"))
    print(f"  {var:<28s}: r={r:+.3f}  p={p:.2e}  {sig}")

# %% [markdown]
# ## 9 · Análisis de proyectos UPME

# %%
print("=" * 60)
print("PROYECTOS UPME — RESUMEN GENERAL")
print("=" * 60)

# Por recurso
rec_summary = (df_upme_all.groupby("recurso")
                           .agg(n_proyectos=("codigo_proyecto","count"),
                                mw_total=("capacidad_mw","sum"),
                                mw_promedio=("capacidad_mw","mean"))
                           .sort_values("mw_total", ascending=False))
print("\nPor recurso energético:")
print(rec_summary.round(1).to_string())

# SOLAR: por región
sol_region = (df_solar.groupby("region_natural")
                       .agg(n_proyectos=("codigo_proyecto","count"),
                            mw_total=("capacidad_mw","sum"),
                            mw_promedio=("capacidad_mw","mean"))
                       .sort_values("mw_total", ascending=False))
print("\nProyectos SOLARES por región natural:")
print(sol_region.round(1).to_string())

# SOLAR: por estado de madurez
sol_estado = (df_solar.groupby("estado")
                       .agg(n=("codigo_proyecto","count"),
                            mw=("capacidad_mw","sum"))
                       .sort_values("mw", ascending=False))
print("\nProyectos SOLARES por estado (fase):")
print(sol_estado.round(1).to_string())

# SOLAR: por departamento
sol_dpto = (df_solar.groupby(["departamento","region_natural"])
                     .agg(n=("codigo_proyecto","count"),
                          mw=("capacidad_mw","sum"))
                     .sort_values("mw", ascending=False)
                     .head(15))
print("\nTop 15 departamentos — capacidad solar instalada/planificada (MW):")
print(sol_dpto.round(1).to_string())

# %% [markdown]
# ## 10 · Visualizaciones UPME — capacidad y distribución

# %%
# ── 10.1 Treemap-style bar: MW por región ────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle("Proyectos UPME — Recurso SOLAR (Fotovoltaico)", fontweight="bold", fontsize=14)

# Por región
ax = axes[0]
sol_region_sorted = sol_region.sort_values("mw_total", ascending=True)
bars = ax.barh(sol_region_sorted.index, sol_region_sorted["mw_total"],
               color=[PALETTE_REGION.get(r,"#888") for r in sol_region_sorted.index])
ax.set_xlabel("MW total planificado")
ax.set_title("Capacidad por región (MW)")
for bar, val in zip(bars, sol_region_sorted["mw_total"]):
    ax.text(bar.get_width() + 20, bar.get_y() + bar.get_height()/2,
            f"{val:,.0f}", va="center", fontsize=9)

# Por estado
ax = axes[1]
colors_estado = {"FASE 1": "#06D6A0", "FASE 2": SOLAR_ORANGE, "FASE 3": GAP_RED}
wedges, texts, autotexts = ax.pie(
    sol_estado["mw"], labels=sol_estado.index,
    colors=[colors_estado.get(e, "#888") for e in sol_estado.index],
    autopct="%1.1f%%", startangle=90,
    wedgeprops=dict(linewidth=1.5, edgecolor="white"))
for at in autotexts:
    at.set_fontsize(10)
ax.set_title("Distribución por fase de madurez (MW)")

# Distribución de capacidad individual (boxplot)
ax = axes[2]
ax.boxplot(df_solar["capacidad_mw"].dropna(),
           vert=True, patch_artist=True,
           boxprops=dict(facecolor=SOLAR_ORANGE, alpha=0.6),
           medianprops=dict(color="#333", lw=2))
ax.set_ylabel("MW por proyecto")
ax.set_title("Distribución individual\npor proyecto (MW)")
ax.set_xticks([])
stats_cap = df_solar["capacidad_mw"].describe()
ax.text(1.3, stats_cap["75%"], f"P75={stats_cap['75%']:.0f}", va="center", fontsize=8)
ax.text(1.3, stats_cap["25%"], f"P25={stats_cap['25%']:.0f}", va="center", fontsize=8)
ax.text(1.3, stats_cap["50%"], f"Med={stats_cap['50%']:.0f}", va="center", fontsize=8)

plt.tight_layout()
plt.savefig("fig07_upme_capacidad.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅  Figura 7 guardada.")

# ── 10.2 Inscripción temporal ─────────────────────────────────────────────────
timeline = (df_solar.groupby(df_solar["fecha_inscripcion"].dt.to_period("M"))
                     .agg(n=("codigo_proyecto","count"),
                          mw=("capacidad_mw","sum"))
                     .reset_index())
timeline["fecha_plot"] = timeline["fecha_inscripcion"].dt.to_timestamp()

fig, ax1 = plt.subplots(figsize=(14, 5))
ax2 = ax1.twinx()
ax1.bar(timeline["fecha_plot"], timeline["mw"], width=25,
        color=SOLAR_ORANGE, alpha=0.7, label="MW inscrito")
ax2.plot(timeline["fecha_plot"], timeline["n"], color=UPME_BLUE,
         marker="o", ms=5, lw=2, label="N° proyectos")
ax1.set_ylabel("MW inscrito por mes", color=SOLAR_ORANGE)
ax2.set_ylabel("Número de proyectos", color=UPME_BLUE)
ax1.tick_params(axis="y", colors=SOLAR_ORANGE)
ax2.tick_params(axis="y", colors=UPME_BLUE)
ax1.set_title("Evolución temporal de inscripciones solares UPME (2024–2026)",
              fontweight="bold")
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2)
plt.tight_layout()
plt.savefig("fig08_upme_timeline.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅  Figura 8 guardada.")

# %% [markdown]
# ## 11 · Construcción del panel de análisis de brecha (GAP)
#
# Para cada municipio del grid NASA se calcula:
# - `irr_media` — irradiancia solar media diaria (kWh/m²/día)
# - `n_proyectos_solar` — número de proyectos UPME solares en ese municipio
# - `mw_solar` — capacidad total UPME solar en ese municipio
#
# Se añade la misma agregación a nivel **departamento** y **región** para
# los análisis de brecha territorial.

# %%
# ── Nivel municipio ───────────────────────────────────────────────────────────
upme_muni = (df_solar.groupby("id_municipio")
                      .agg(n_proyectos_solar=("codigo_proyecto","count"),
                           mw_solar=("capacidad_mw","sum"))
                      .reset_index())

panel_muni = irr_muni.merge(upme_muni, on="id_municipio", how="left")
panel_muni["n_proyectos_solar"] = panel_muni["n_proyectos_solar"].fillna(0).astype(int)
panel_muni["mw_solar"]          = panel_muni["mw_solar"].fillna(0)
panel_muni["tiene_solar"]       = (panel_muni["n_proyectos_solar"] > 0).astype(int)

# ── Nivel región ──────────────────────────────────────────────────────────────
irr_region = (df_nasa.groupby("region_natural")
                      .agg(irr_media=("irradiancia_global","mean"),
                           irr_std=("irradiancia_global","std"))
                      .reset_index())

upme_region = (df_solar.groupby("region_natural")
                        .agg(n_proyectos=("codigo_proyecto","count"),
                             mw_total=("capacidad_mw","sum"),
                             mw_promedio=("capacidad_mw","mean"))
                        .reset_index())

panel_region = irr_region.merge(upme_region, on="region_natural", how="left")
panel_region = panel_region.sort_values("irr_media", ascending=False)

# Normalizar para índice de brecha (Z-score inverso de MW vs irradiancia)
from sklearn.preprocessing import MinMaxScaler   # sólo para escalar 0-1
try:
    from sklearn.preprocessing import MinMaxScaler
    scaler = MinMaxScaler()
    panel_region["irr_norm"] = scaler.fit_transform(panel_region[["irr_media"]])
    # MW nulo para regiones sin proyectos = 0
    panel_region["mw_total"] = panel_region["mw_total"].fillna(0)
    panel_region["mw_norm"]  = scaler.fit_transform(panel_region[["mw_total"]])
    panel_region["gap_score"] = panel_region["irr_norm"] - panel_region["mw_norm"]
except ImportError:
    # Fallback sin sklearn
    for col in ["irr_media","mw_total"]:
        panel_region[col+"_norm"] = (
            (panel_region[col] - panel_region[col].min()) /
            (panel_region[col].max() - panel_region[col].min() + 1e-9)
        )
    panel_region.rename(columns={"irr_media_norm":"irr_norm",
                                  "mw_total_norm":"mw_norm"}, inplace=True)
    panel_region["gap_score"] = panel_region["irr_norm"] - panel_region["mw_norm"]

panel_region["mw_total"] = panel_region["mw_total"].fillna(0)
panel_region["n_proyectos"] = panel_region["n_proyectos"].fillna(0).astype(int)

print("Panel regional construido:")
print(panel_region[["region_natural","irr_media","mw_total","n_proyectos","gap_score"]]
      .round(3).to_string(index=False))

# %% [markdown]
# ## 12 · Análisis de correlación irradiancia ↔ proyectos UPME

# %%
# ── 12.1 Nivel municipio (todos los que tienen datos NASA) ───────────────────
print("=" * 60)
print("CORRELACIÓN IRRADIANCIA ↔ PROYECTOS SOLARES (nivel municipio)")
print("=" * 60)

sub = panel_muni[["irr_media","n_proyectos_solar","mw_solar"]].dropna()

for metric, label in [("n_proyectos_solar","N° proyectos"), ("mw_solar","MW planificados")]:
    r_p, p_p  = pearsonr(sub["irr_media"], sub[metric])
    r_s, p_s  = spearmanr(sub["irr_media"], sub[metric])
    r_k, p_k  = kendalltau(sub["irr_media"], sub[metric])
    print(f"\n  [{label}]")
    print(f"    Pearson  r={r_p:+.3f}  p={p_p:.4f}  {'sig***' if p_p<0.001 else ('sig**' if p_p<0.01 else ('sig*' if p_p<0.05 else 'NO sig'))}")
    print(f"    Spearman ρ={r_s:+.3f}  p={p_s:.4f}")
    print(f"    Kendall  τ={r_k:+.3f}  p={p_k:.4f}")

# ── 12.2 Scatter irr_media vs mw_solar con regresión ─────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle("Correlación: Irradiancia media vs. Proyectos UPME Solar\n(nivel municipio)",
             fontweight="bold")

for ax, (metric, label) in zip(axes,
    [("n_proyectos_solar","N° proyectos fotovoltaicos"),
     ("mw_solar","Capacidad MW planificada")]):

    colors_dot = [PALETTE_REGION.get(r,"#888") for r in panel_muni["region_natural"]]
    ax.scatter(panel_muni["irr_media"], panel_muni[metric],
               c=colors_dot, alpha=0.75, s=60, edgecolors="white", lw=0.5)

    # OLS line
    sub2 = panel_muni[["irr_media",metric]].dropna()
    m, b, r, p, se = stats.linregress(sub2["irr_media"], sub2[metric])
    x_range = np.linspace(sub2["irr_media"].min(), sub2["irr_media"].max(), 100)
    ax.plot(x_range, m*x_range + b, color="#333", lw=1.5, ls="--",
            label=f"OLS: y={m:.1f}x{b:+.1f}\nr={r:.3f}, p={p:.3f}")

    ax.set_xlabel("Irradiancia media diaria (kWh/m²/día)")
    ax.set_ylabel(label)
    ax.set_title(label)
    ax.legend(fontsize=9)

    # Anotar municipios outlier
    threshold = panel_muni[metric].quantile(0.90)
    for _, row in panel_muni[panel_muni[metric] >= threshold].iterrows():
        ax.annotate(row["municipio"].title(),
                    xy=(row["irr_media"], row[metric]),
                    xytext=(4, 4), textcoords="offset points",
                    fontsize=7, color="#333")

# Leyenda de regiones
legend_patches = [mpatches.Patch(color=c, label=r) for r, c in PALETTE_REGION.items()
                  if r in panel_muni["region_natural"].values]
fig.legend(handles=legend_patches, loc="lower center", ncol=6,
           fontsize=9, title="Región natural", bbox_to_anchor=(0.5, -0.05))

plt.tight_layout()
plt.savefig("fig09_correlacion_scatter.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅  Figura 9 guardada.")

# %% [markdown]
# ## 13 · Análisis de brecha (GAP) recurso ↔ planificación

# %%
# ── 13.1 Brecha a nivel regional ─────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(11, 6))
x = np.arange(len(panel_region))
width = 0.38

bars1 = ax.bar(x - width/2, panel_region["irr_norm"]*100,
               width, color=SOLAR_ORANGE, alpha=0.8, label="Irradiancia (normalizada 0-100)")
bars2 = ax.bar(x + width/2, panel_region["mw_norm"]*100,
               width, color=UPME_BLUE,    alpha=0.8, label="MW UPME solar (normalizado 0-100)")

ax.set_xticks(x)
ax.set_xticklabels(panel_region["region_natural"], fontsize=11)
ax.set_ylabel("Índice normalizado (0 = mín, 100 = máx)")
ax.set_title("Brecha entre recurso solar disponible y planificación UPME\n(índices normalizados por región)",
             fontweight="bold")
ax.legend()

# Flechas de brecha
for i, (_, row) in enumerate(panel_region.iterrows()):
    y_irr = row["irr_norm"] * 100
    y_mw  = row["mw_norm"]  * 100
    if abs(y_irr - y_mw) > 3:
        ax.annotate("", xy=(i + width/2, y_mw), xytext=(i - width/2, y_irr),
                    arrowprops=dict(arrowstyle="->", color=GAP_RED, lw=1.5))
        mid = (y_irr + y_mw) / 2
        gap_val = row.get("gap_score", 0)
        ax.text(i + 0.02, mid, f"gap={gap_val:.2f}",
                fontsize=8, color=GAP_RED, va="center")

plt.tight_layout()
plt.savefig("fig10_brecha_regional.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅  Figura 10 guardada.")

# ── 13.2 Mapa de calor: municipios de alta irradiancia sin proyectos ──────────
high_irr_threshold = panel_muni["irr_media"].quantile(0.70)   # top 30 %
gap_munis = panel_muni[
    (panel_muni["irr_media"] >= high_irr_threshold) &
    (panel_muni["n_proyectos_solar"] == 0)
].sort_values("irr_media", ascending=False)

print(f"\nMunicipios en top 30% irradiancia (≥ {high_irr_threshold:.3f} kWh/m²/día) SIN proyectos UPME solar:")
print(f"Total: {len(gap_munis)} de {(panel_muni['irr_media'] >= high_irr_threshold).sum()} municipios de alta irradiancia\n")
print(gap_munis[["municipio","departamento","region_natural","irr_media","n_proyectos_solar"]]
      .to_string(index=False))

# ── 13.3 Scatter Irradiancia vs MW coloreado por gap ─────────────────────────
fig, ax = plt.subplots(figsize=(12, 7))

# Color por región, tamaño por irradiancia
for region in panel_muni["region_natural"].unique():
    sub_r = panel_muni[panel_muni["region_natural"] == region]
    ax.scatter(sub_r["irr_media"], sub_r["mw_solar"],
               c=PALETTE_REGION.get(region, "#888"),
               s=sub_r["irr_media"] * 20, alpha=0.7,
               edgecolors="white", lw=0.5, label=region)

# Destacar gap munis con borde rojo
ax.scatter(gap_munis["irr_media"], gap_munis["mw_solar"],
           s=gap_munis["irr_media"] * 20 + 10,
           facecolors="none", edgecolors=GAP_RED, lw=2.2, zorder=5,
           label="Alta irradiancia SIN proyecto UPME")

# Anotar top gap
for _, row in gap_munis.head(8).iterrows():
    ax.annotate(f"{row['municipio'].title()}",
                xy=(row["irr_media"], row["mw_solar"]),
                xytext=(6, 3), textcoords="offset points",
                fontsize=7.5, color=GAP_RED,
                arrowprops=dict(arrowstyle="-", color=GAP_RED, lw=0.5))

ax.axvline(high_irr_threshold, color="#aaa", ls=":", lw=1.3,
           label=f"Umbral top 30% irr ({high_irr_threshold:.2f})")
ax.set_xlabel("Irradiancia media diaria (kWh/m²/día)")
ax.set_ylabel("Capacidad MW solar UPME en el municipio")
ax.set_title("Irradiancia vs. Capacidad UPME Solar — análisis de brecha territorial",
             fontweight="bold")
ax.legend(loc="upper left", fontsize=8)
plt.tight_layout()
plt.savefig("fig11_brecha_scatter.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅  Figura 11 guardada.")

# %% [markdown]
# ## 14 · Análisis multivariado — perfil climático de municipios con/sin proyectos UPME

# %%
# Enriquecer panel_muni con promedios climatológicos
clima_muni = (df_nasa.groupby("id_municipio")
                      .agg(irr_media=("irradiancia_global","mean"),
                           temp_media=("temperatura_2m","mean"),
                           hr_media=("humedad_relativa","mean"),
                           ws_media=("velocidad_viento","mean"),
                           prec_media=("precipitacion","mean"))
                      .reset_index())

panel_muni2 = panel_muni.merge(clima_muni, on="id_municipio", suffixes=("","_dup"))
panel_muni2["grupo"] = panel_muni2["tiene_solar"].map({1:"Con proyecto UPME",
                                                         0:"Sin proyecto UPME"})

vars_clima = ["irr_media","temp_media","hr_media","ws_media","prec_media"]
labels_clima = ["Irradiancia\n(kWh/m²/d)","Temperatura\n(°C)",
                "Humedad rel.\n(%)","Viento\n(m/s)","Precipitación\n(mm/d)"]

# Comparación de medias (t-test)
print("Diferencias estadísticas entre municipios CON y SIN proyecto UPME solar:")
con  = panel_muni2[panel_muni2["tiene_solar"] == 1]
sin_ = panel_muni2[panel_muni2["tiene_solar"] == 0]
for var, lbl in zip(vars_clima, labels_clima):
    t, p = stats.ttest_ind(con[var].dropna(), sin_[var].dropna())
    sig = "***" if p < 0.001 else ("**" if p < 0.01 else ("*" if p < 0.05 else "ns"))
    print(f"  {lbl.replace(chr(10),' '):<30s}: "
          f"CON={con[var].mean():.3f}  SIN={sin_[var].mean():.3f}  "
          f"t={t:.2f}  p={p:.4f}  {sig}")

# Violin plot comparativo
fig, axes = plt.subplots(1, 5, figsize=(17, 6))
fig.suptitle("Perfil climático: Municipios CON vs. SIN proyectos UPME solar",
             fontweight="bold")

palette_grupo = {"Con proyecto UPME": SOLAR_ORANGE, "Sin proyecto UPME": "#AAAAAA"}

for ax, var, lbl in zip(axes, vars_clima, labels_clima):
    sns.violinplot(data=panel_muni2, x="grupo", y=var,
                   palette=palette_grupo, ax=ax, inner="box", cut=0)
    ax.set_title(lbl, fontsize=9)
    ax.set_xlabel("")
    ax.set_ylabel("")
    ax.tick_params(axis="x", labelsize=8, rotation=15)

plt.tight_layout()
plt.savefig("fig12_perfil_climatico_upme.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅  Figura 12 guardada.")

# %% [markdown]
# ## 15 · Análisis de componentes principales (PCA) — variables climáticas

# %%
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

X = panel_muni2[vars_clima].dropna()
idx = X.index
scaler_pca = StandardScaler()
X_scaled = scaler_pca.fit_transform(X)

pca = PCA(n_components=2, random_state=42)
comps = pca.fit_transform(X_scaled)
df_pca = pd.DataFrame(comps, columns=["PC1","PC2"], index=idx)
df_pca = df_pca.join(panel_muni2[["municipio","region_natural","tiene_solar","mw_solar"]])

exp_var = pca.explained_variance_ratio_
print(f"Varianza explicada → PC1: {exp_var[0]:.1%}  PC2: {exp_var[1]:.1%}  "
      f"Total: {exp_var.sum():.1%}")
print("\nCargas de componentes:")
loadings = pd.DataFrame(pca.components_.T, index=vars_clima, columns=["PC1","PC2"])
print(loadings.round(3).to_string())

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle("PCA — Espacio climático de municipios del grid NASA",
             fontweight="bold")

# Biplot coloreado por región
ax = axes[0]
for region in df_pca["region_natural"].unique():
    sub = df_pca[df_pca["region_natural"] == region]
    ax.scatter(sub["PC1"], sub["PC2"],
               c=PALETTE_REGION.get(region,"#888"),
               s=50, alpha=0.75, edgecolors="white", lw=0.4, label=region)
# Vectores de carga
scale = 2.5
for i, (var, lbl) in enumerate(zip(vars_clima, labels_clima)):
    ax.arrow(0, 0, loadings.loc[var,"PC1"]*scale, loadings.loc[var,"PC2"]*scale,
             head_width=0.07, head_length=0.05, fc="#333", ec="#333")
    ax.text(loadings.loc[var,"PC1"]*scale*1.15, loadings.loc[var,"PC2"]*scale*1.15,
            lbl.replace("\n"," "), fontsize=7.5, ha="center", color="#333")
ax.set_xlabel(f"PC1 ({exp_var[0]:.1%})")
ax.set_ylabel(f"PC2 ({exp_var[1]:.1%})")
ax.set_title("Por región natural")
ax.legend(fontsize=8)
ax.axhline(0, color="#ccc", lw=0.8); ax.axvline(0, color="#ccc", lw=0.8)

# Coloreado por presencia UPME
ax = axes[1]
cols_upme = df_pca["tiene_solar"].map({1: SOLAR_ORANGE, 0: "#BBBBBB"})
sizes_upme = df_pca["mw_solar"].fillna(0).apply(lambda v: max(30, v*0.3))
scatter = ax.scatter(df_pca["PC1"], df_pca["PC2"],
                     c=cols_upme, s=sizes_upme, alpha=0.75,
                     edgecolors="white", lw=0.4)
ax.set_xlabel(f"PC1 ({exp_var[0]:.1%})")
ax.set_ylabel(f"PC2 ({exp_var[1]:.1%})")
ax.set_title("Por presencia de proyecto UPME solar\n(tamaño ∝ MW)")
legend_el = [
    mpatches.Patch(color=SOLAR_ORANGE, label="Con proyecto UPME"),
    mpatches.Patch(color="#BBBBBB",    label="Sin proyecto UPME"),
]
ax.legend(handles=legend_el, fontsize=9)
ax.axhline(0, color="#ccc", lw=0.8); ax.axvline(0, color="#ccc", lw=0.8)

plt.tight_layout()
plt.savefig("fig13_pca_climatico.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅  Figura 13 guardada.")

# %% [markdown]
# ## 16 · Mapa de dispersión geoespacial (latitud/longitud)

# %%
# ── Panel: scatter lat/lon coloreado por irradiancia + proyectos ──────────────
fig, axes = plt.subplots(1, 2, figsize=(16, 9))
fig.suptitle("Distribución geoespacial: Irradiancia solar vs. Proyectos UPME",
             fontweight="bold", fontsize=14)

# Irradiancia
ax = axes[0]
sc = ax.scatter(panel_muni2["longitud"], panel_muni2["latitud"],
                c=panel_muni2["irr_media"],
                cmap="YlOrRd", s=80, alpha=0.85,
                edgecolors="#555", lw=0.5, vmin=3.7, vmax=5.8)
plt.colorbar(sc, ax=ax, label="Irradiancia media (kWh/m²/día)", shrink=0.7)
ax.set_xlabel("Longitud"); ax.set_ylabel("Latitud")
ax.set_title("Irradiancia media 2024–2026\npor municipio del grid NASA")
# Anotar top-5
for _, row in irr_muni.head(5).iterrows():
    info = panel_muni2[panel_muni2["id_municipio"] == row["id_municipio"]]
    if not info.empty:
        ax.annotate(row["municipio"].title(),
                    xy=(info.iloc[0]["longitud"], info.iloc[0]["latitud"]),
                    fontsize=7, color="#333",
                    xytext=(4,4), textcoords="offset points")

# Proyectos UPME solar
ax = axes[1]
# Municipios sin proyecto (fondo)
sin_proj = panel_muni2[panel_muni2["mw_solar"] == 0]
ax.scatter(sin_proj["longitud"], sin_proj["latitud"],
           c=sin_proj["irr_media"], cmap="YlOrRd",
           s=40, alpha=0.4, edgecolors="#aaa", lw=0.3, vmin=3.7, vmax=5.8,
           label="Sin proyecto UPME")
# Municipios con proyecto
con_proj = panel_muni2[panel_muni2["mw_solar"] > 0]
sc2 = ax.scatter(con_proj["longitud"], con_proj["latitud"],
                 c=con_proj["irr_media"], cmap="YlOrRd",
                 s=con_proj["mw_solar"].clip(upper=800)/5 + 50,
                 alpha=0.9, edgecolors=UPME_BLUE, lw=2,
                 vmin=3.7, vmax=5.8, label="Con proyecto UPME (tamaño ∝ MW)")
plt.colorbar(sc2, ax=ax, label="Irradiancia media (kWh/m²/día)", shrink=0.7)
# Municipios gap (alta irr, sin proyecto)
ax.scatter(gap_munis["longitud"], gap_munis["latitud"],
           s=80, facecolors="none", edgecolors=GAP_RED, lw=2, zorder=6,
           label="BRECHA: alta irr. sin proyecto")
for _, row in gap_munis.head(6).iterrows():
    ax.annotate(row["municipio"].title(),
                xy=(row["longitud"], row["latitud"]),
                fontsize=6.5, color=GAP_RED,
                xytext=(4,3), textcoords="offset points")
ax.set_xlabel("Longitud"); ax.set_ylabel("Latitud")
ax.set_title("Proyectos UPME solar sobre el mapa de irradiancia\n(brecha en rojo)")
ax.legend(loc="lower left", fontsize=7.5)

plt.tight_layout()
plt.savefig("fig14_mapa_geoespacial.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅  Figura 14 guardada.")

# %% [markdown]
# ## 17 · Tabla resumen ejecutiva — índice de brecha por región

# %%
# Construir tabla ejecutiva
exec_table = panel_region[["region_natural","irr_media","irr_std",
                             "n_proyectos","mw_total","gap_score"]].copy()
exec_table.columns = ["Región","Irr. media\n(kWh/m²/d)","Irr. std",
                       "N° proy. solar","MW UPME solar","Gap score\n(irr-MW norm)"]
exec_table = exec_table.sort_values("Gap score\n(irr-MW norm)", ascending=False)
exec_table = exec_table.reset_index(drop=True)
exec_table.index += 1

# Gap por departamento
exec_dpto = (panel_muni2.groupby(["departamento","region_natural"])
                          .agg(irr_media=("irr_media","mean"),
                               n_proyectos=("n_proyectos_solar","sum"),
                               mw_solar=("mw_solar","sum"),
                               n_munis=("id_municipio","count"),
                               n_munis_solar=("tiene_solar","sum"))
                          .reset_index()
                          .sort_values("irr_media", ascending=False))
exec_dpto["pct_munis_con_solar"] = (
    100 * exec_dpto["n_munis_solar"] / exec_dpto["n_munis"]).round(1)

print("=" * 70)
print("TABLA EJECUTIVA — ÍNDICE DE BRECHA POR REGIÓN NATURAL")
print("=" * 70)
print(exec_table.round(3).to_string())

print("\n\nRESUMEN POR DEPARTAMENTO (irradiancia ≥ 5.0 kWh/m²/día):")
top_dpto = exec_dpto[exec_dpto["irr_media"] >= 5.0]
print(top_dpto[["departamento","region_natural","irr_media","n_proyectos","mw_solar","pct_munis_con_solar"]]
      .round(3).to_string(index=False))

# Gráfico de tabla ejecutiva como figura
fig, ax = plt.subplots(figsize=(14, 3.5))
ax.axis("off")
tbl_data = exec_table.map(lambda v: f"{v:.2f}" if isinstance(v, float) else str(v))
table = ax.table(
    cellText=tbl_data.values,
    colLabels=exec_table.columns,
    cellLoc="center", loc="center",
    bbox=[0, 0, 1, 1]
)
table.auto_set_font_size(False)
table.set_fontsize(10)
for key, cell in table.get_celld().items():
    if key[0] == 0:
        cell.set_facecolor(UPME_BLUE)
        cell.set_text_props(color="white", fontweight="bold")
    elif exec_table.iloc[key[0]-1]["Gap score\n(irr-MW norm)"] > 0.3:
        cell.set_facecolor("#FFF3CD")  # amarillo suave → brecha alta
    else:
        cell.set_facecolor("#FFFFFF")
ax.set_title("Índice de brecha recurso solar ↔ planificación UPME por región",
             fontweight="bold", y=1.02)
plt.tight_layout()
plt.savefig("fig15_tabla_ejecutiva.png", dpi=150, bbox_inches="tight")
plt.show()
print("✅  Figura 15 guardada.")

# %% [markdown]
# ## 18 · Resumen estadístico consolidado y conclusiones

# %%
print("""
╔══════════════════════════════════════════════════════════════════════════════╗
║        RESUMEN DE HALLAZGOS — EDA Potencial Solar vs. UPME Colombia        ║
╠══════════════════════════════════════════════════════════════════════════════╣
║                                                                              ║
║  COBERTURA DE DATOS                                                          ║
║  · 88 municipios  ·  860 días  ·  74 984 mediciones diarias NASA POWER      ║
║  · 72 proyectos fotovoltaicos UPME  ·  7 155 MW planificados                ║
║                                                                              ║
║  IRRADIANCIA SOLAR (ALLSKY_SFC_SW_DWN)                                      ║
║  · Media nacional: 4.81 kWh/m²/día                                          ║
║  · Mayor potencial: Región Caribe (5.22) e Insular (5.43)                   ║
║  · Menor potencial: Región Pacífica (4.07) y Amazonia (4.50)                ║
║  · Top municipios: Sta. Marta (5.76), Riohacha (5.61), Barranquilla (5.34) ║
║  · Estacionalidad: pico en ene–feb, mínimo en nov; correlación negativa     ║
║    con precipitación (r ≈ −0.31, p < 0.001)                                 ║
║                                                                              ║
║  PROYECTOS UPME SOLAR                                                        ║
║  · 73 % de los MW: Caribe (3 826) y Andina (2 640)                          ║
║  · Departamento líder: Córdoba (18 proyectos · 2 018 MW)                    ║
║  · 87 % de la capacidad en FASE 2; sólo 4 % en FASE 3 (listo para operar)  ║
║  · Crecimiento acelerado 2026: 3 619 MW inscritos vs 1 520 en 2024          ║
║                                                                              ║
║  CORRELACIÓN IRRADIANCIA ↔ PROYECTOS UPME                                   ║
║  · Correlación positiva pero MODERADA a nivel municipal                      ║
║    (Spearman ρ ≈ +0.25–0.30 para N° proyectos y MW)                         ║
║  · Regiones de alta irradiancia (Caribe) SÍ concentran más MW UPME          ║
║  · Sin embargo, la planificación responde también a infraestructura de       ║
║    transmisión, tierra y conectividad —no sólo al recurso solar—             ║
║                                                                              ║
║  BRECHA RECURSO ↔ PLANIFICACIÓN (GAP)                                        ║
║  · Regiones con MAYOR brecha (alta irradiancia, baja planificación):         ║
║    1. Orinoquia: irr 4.73 — sólo 690 MW (gap score alto)                   ║
║    2. Insular:   irr 5.43 — 0 MW UPME registrados                          ║
║    3. Amazonia:  irr 4.50 — 0 MW UPME registrados                          ║
║  · Municipios de ALTA irradiancia sin proyecto UPME (top brecha):            ║
║    Santa Marta, San Andrés, Barranquilla, Tubara, Fundación,                 ║
║    Sincelejo, Puerto Carreño, La Jagua de Ibirico                            ║
║  · La Andina presenta una lógica inversa: alta planificación en zonas        ║
║    de irradiancia media (Colombia interior), impulsada por mercados          ║
║    industriales y capacidad de transmisión existente.                        ║
║                                                                              ║
╚══════════════════════════════════════════════════════════════════════════════╝
""")

print("Figuras generadas:")
figs = [
    "fig01_distribuciones_region.png",
    "fig02_ranking_municipal_irradiancia.png",
    "fig03_boxplot_region.png",
    "fig04_estacionalidad.png",
    "fig05_estacionalidad_region.png",
    "fig06_correlacion_nasa.png",
    "fig07_upme_capacidad.png",
    "fig08_upme_timeline.png",
    "fig09_correlacion_scatter.png",
    "fig10_brecha_regional.png",
    "fig11_brecha_scatter.png",
    "fig12_perfil_climatico_upme.png",
    "fig13_pca_climatico.png",
    "fig14_mapa_geoespacial.png",
    "fig15_tabla_ejecutiva.png",
]
for f in figs:
    print(f"  · {f}")
print(f"\nTotal: {len(figs)} figuras  ·  análisis completado ✅")
```
